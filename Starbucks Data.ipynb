```python
df = pd.DataFrame(starbucks)

# Sort by calories and select the top 10 highest calorie drinks
low_calorie_drinks = df.sort_values(by='Calories', ascending=True).head(40)

# Plotting the bar graph
plt.figure(figsize=(12, 8))
plt.bar(low_calorie_drinks['Beverage'], low_calorie_drinks['Calories'], color='darkred')
plt.xlabel('Drink')
plt.ylabel('Calories')
plt.title('Top 15 Lowest Calorie Starbucks Drinks')
plt.xticks(rotation=90)  # Rotate the drink names to prevent overlapping
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data representing the annual income distribution of Starbucks customers
income_distribution = {
    '$30K or less': 20,
    '$31K-$60K': 25,
    '$61K-$90K': 30,
    '$91K-$120K': 15,
    '$121K or more': 10
}

# Labels and sizes for the pie chart
labels = income_distribution.keys()
sizes = income_distribution.values()

# Define Seaborn color palette for the pie chart
colors = sns.color_palette('pastel')

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})

# Adding a title to the chart
plt.title('Annual Income Distribution of Starbucks Customers', fontsize=16)

# Display the pie chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
starbucks_customers = pd.read_csv('Starbucks satisfactory survey.csv')

# Count the occurrences of each gender
gender_counts = starbucks_customers['1. Your Gender'].value_counts()

# Plotting the data
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Starbucks Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()
```
